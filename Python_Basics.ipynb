```python
Hailstone Sequence Algorithm
Take a number (says ‘n’) as an input.
Print the value of ‘n’.
while n != 1
a) Print the value of ‘n’.
b) If ‘n’ is odd, calculate the next number as n*3+1.
c) If ‘n’ is even, calculate the next number as n/2.
Example:
Input : 10
Output (Hailstone Sequence): 10 -> 5 -> 16 -> 8 -> 4 -> 2 ->1
Input : 17
Output (Hailstone Sequence): 17 -> 52 -> 26 -> 13 -> 40 -> 20 ->10 -> 5 -> 16 -> 8 -> 4 -> 2 -> 1
```

```python
n = int(input())
while n!=1:
    print(n,end=' -> ')
    if n%2!=0:
        n = n*3+1
    else:
        n = n//2
print(n)
```

```python
#que1
n = input("Enter Number: ")
even =0
odd =0
zero =0
for ch in n:
  if int(ch)%2==0:
    even+=1
  elif int(ch)==0:
    zero+=1
  else:
    odd+=1
print(f"Even :{even} , odd:{odd} , zeros: {zero}")
```

```python
#que3
n = int(input("""-- Enter 1 to ADD BOOK INFORMATION
-- Enter 2 to DISPLAY BOOK INFORMATION
-- Enter 3 to LIST ALL BOOKS OF GIVEN AUTHOR
-- Enter 4 to LIST THE COUNT OF BOOKS IN THE LIBRARY
-- enter 5 to EXIT
"""))
list_of_books = {}
if n==1:
  title = input("Enter Title: ")
  author = input("Enter Author: ")
  cost = int(input("Enter Cost: "))
  list_of_books[title] = [author,cost]
  print("Added Successfully",list_of_books)
elif n==2:
  title_input = input("Enter Title: ")
  if title_input in list_of_books:
    author, cost = list_of_books[title_input]
    print(f"Book Title: {title_input}")
    print(f"Author: {author}")
    print(f"Cost: {cost}")
  else:
    print("Not Found")
elif n==3:
  author_input = input("Enter Author: ")
  found_books = False
  print(f"Books by {author_input}:")
  for title, details in list_of_books.items():
    if details[0] == author_input:
      print(f"  - Title: {title}, Cost: {details[1]}")
      found_books = True
  if not found_books:
    print("  No books found by this author.")
elif n==4:
  print(f"Total number of books in the library: {len(list_of_books)}")
elif n==5:
  print("Exiting program.")
else:
  print("Invalid option. Please enter a number between 1 and 5.")
```
